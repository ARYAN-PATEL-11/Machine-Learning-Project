```python
#Importing python libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Loading 1st dataset into jupyter notebook
df = pd.read_csv('adult+stretch.data',index_col = False , names = ['COLOR','SIZE','ACT','AGE','INFLATED'])
#Displaying 1st dataset
df
```

```python
#Loading 2nd dataset into jupyter notebook
df2 = pd.read_csv('adult-stretch.data',index_col = False , names = ['COLOR','SIZE','ACT','AGE','INFLATED'])
#Displaying 2nd dataset
df2
```

```python
#Loading 3rd dataset into jupyter notebook
df3 = pd.read_csv('yellow-small.data',index_col = False , names = ['COLOR','SIZE','ACT','AGE','INFLATED'])
#Displaying 3rd dataset
df3
```

```python
#Loading 4th dataset into jupyter notebook
df4 = pd.read_csv('yellow-smalladult-stretch.data',index_col = False , names = ['COLOR','SIZE','ACT','AGE','INFLATED'])
#Displaying 4th dataset
df4
```

```python
#for Qualitative data dtype can be of type "object" also
object_feature_mask = df7.dtypes=='object' 
object_feature_mask
```

```python
#Displaying attributes of dataset as a list
object_cols = df7.columns[object_feature_mask].tolist()
object_cols
```

```python
#importing scikit learn library
import sklearn
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()
# applying le on categorical feature columns for converting object data into numeric
df7[object_cols] = df7[object_cols].apply(lambda col: le.fit_transform(col))
df7[object_cols]
```

Since skewness is very close to zero for Size so it Normally distributed

```python
#we will begin by applying log transformation to skewed numeric features
from scipy.stats import skew
num_data = df7.dtypes[df7.dtypes!="object"].index
skew_data = df7[num_data].apply(lambda x: skew(x.dropna()))
skew_data = skew_data[skew_data > 0.75]
skew_data = skew_data.index
skew_data
df7[skew_data] = np.log1p(df7[skew_data])
```

```python
#drop variables noted in EDA section
drop_me = ['COLOR']
df7 = df7.drop(drop_me, axis=1)
```

```python
df7.head()
```

```python
#split concatenated data into train and test dataframes
from sklearn.model_selection import train_test_split
y = df7['INFLATED']
X = df7.loc[:,df7.columns!='class']
print(X.shape)
```

```python
#Declaring the train and test variables
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3 , random_state=42)
```

```python
#Printing the train and test variables
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Build the Model
 
     We will attempt to apply the following models:
           Logistic Regression
           Decision Tree
           Random Forests

-->Logistic Regression

```python
#importing Logistic Regression and using it in our model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver = "liblinear" , multi_class = "ovr") 
lr.fit(X_train , y_train)
lr_predict = lr.predict(X_test)
lr_predict
```

```python
#importing Decision Tree Classifier and using it in our model
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train , y_train)
dt_predict = dt.predict(X_test)
dt_predict
```

```python
#importing Random Forest Classifier and using it in our model
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(criterion = "entropy")
rf.fit(X_train , y_train)
rf_predict = rf.predict(X_test)
```

```python
rf_predict
```

```python
# Train score & Test score of Logistic Regression
from sklearn.metrics import accuracy_score 
print("Train Accuracy of Logistic Regression", lr.score (X_train, y_train)*100) 
print("Accuracy (Test) score of Logistic Regression", lr.score (X_test, y_test)*100) 
print("Accuracy (Test) score of Logistic Regression", accuracy_score(y_test, lr_predict)*100)
```

```python
#Train score & Test score of Decision Tree 
print("Train Accuracy of Decesion Tree", dt.score(X_train,y_train)*100) 
print("Accuracy (Test) score of Decesion Tree", dt.score (X_test, y_test)*100) 
print("Accuracy score of Decesion Tree", accuracy_score(y_test,dt_predict)*100)
```

```python
# Train score & Test score of Random Forest 
print ("Train Accuracy of Random Forest", rf.score (X_train,y_train)*100) 
print("Accuracy (Test) score of Random Forest",rf.score(X_test, y_test)*100) 
print("Accuracy score of Random Forest", accuracy_score(y_test, rf_predict)*100)
```
